```python
def cleanNb(x):
    try:
        return int(x/1000)
    except:
        return 'error:' + str(x)
 
```

```python
pdMaires2014 = pd.read_csv('./data/2014/maires-17-06-2014.zip', compression='zip', header=0, sep=';', quotechar='"')
pdMaires2014.codeinsee = pd.to_numeric(pdMaires2014.codeinsee, errors='coerce', downcast='integer') 
pdMaires2014["coddpt"] = pdMaires2014.codeinsee.apply(lambda x: cleanNb(x)) 
pdMaires2014["codsubcom"] = pdMaires2014.codeinsee - 1000*pdMaires2014.coddpt
pdMaires2014= pdMaires2014.rename(columns={"coddpt": "Code du département","codsubcom": "Code de la commune", "nompsn": "NomElu2014","prepsn": "PrenomElu2014"})
pdMaires2014["Nom"] = pdMaires2014.NomElu2014

print(len(pdMaires2014))
pdMaires2014
```

## Regles

### Less than 1000 inhabitants

Obtiennent un siège au conseil municipal au premier tour les candidats remplissant une double condition : avoir obtenu la majorité absolue des suffrages exprimés et recueilli au moins un quart des voix des électeurs inscrits. 

```python
pdSmall_reelected = pdSmall[(pdSmall['% Voix/Ins'] >= 25.0) & (pdSmall["% Voix/Exp"] > 50.0)]
pdSmall_simple = pdSmall_reelected[["Code du département","Code de la commune","Nom","Prénom",'% Voix/Ins',"% Voix/Exp"]]
pdSmall_simple["type"] = "small"
pdSmall_simple
pdSmall_simple.head()
```

### More than 1000 inhabitants

Au premier tour, la liste qui obtient la majorité absolue des suffrages exprimés reçoit un nombre de sièges égal à la moitié des sièges à pourvoir. Les autres sièges sont répartis à la représentation proportionnelle à la plus forte moyenne entre toutes les listes ayant obtenu plus de 5% des suffrages exprimés, en fonction du nombre de suffrage obtenus.

__attention, il n'y a pas le nom de la liste / son orientation__

```python
pdLarge_reelected = pdLarge[(pdLarge["% Voix/Exp"] > 50.0)]
pdLarge_simple = pdLarge_reelected[["Code du département","Code de la commune","Nom","Prénom",'% Voix/Ins',"% Voix/Exp"]]
pdLarge_simple["type"] = "large"
pdLarge_simple
```

```python
k1 = set(pdLarge_reelected.keys().tolist())
k2 = set(pdSmall_reelected.keys().tolist())
toremove = list(k2 - k1)
pdSmall_reelected = pdSmall_reelected.drop(columns=toremove)
```

```python
pdElus1er2020 = pd.concat([pdLarge_simple,pdSmall_simple])
pdElus1er2020.to_excel("./data/pdReElectedAll.xls")
pdElus1er2020
```

```python
LARGE = pd2014[["Code de la commune","Code du département","NomElu2014", "PrenomElu2014","codnualisteinitiale"]].merge(pdLarge_simple,on=["Code de la commune","Code du département"],how="right")
LARGE = LARGE.dropna(subset=['NomElu2014', 'PrenomElu2014','Nom', 'Prénom'],thresh=2)
print(len(LARGE))
LARGE.tail()
```

```python
ReElectedLarge = []
CityLarge = []
for d in pdLarge_simple["Code du département"].unique():
    #print(d)
    dpt = 0
    for c in pdLarge_simple[pdLarge_simple["Code du département"] == d]["Code de la commune"].unique():
        if (c in pd2014["Code de la commune"].tolist()) and (d in pd2014["Code du département"].tolist()):
            pdtempA = pd2014[(pd2014["Code de la commune"] == c)&(pd2014["Code du département"] == d)]
            pdtempB = pdLarge_simple[(pdLarge_simple["Code de la commune"] == c) &(pdLarge_simple["Code du département"] == d)]
            #print(d,c)
            if len(pdtempA) and len(pdtempB):
                if pdtempB.Nom.unique()[0] in pdtempA.NomElu2014.unique().tolist():
                    #print(pdtempB.Nom.unique()[0],pdtempA.NomElu2014.unique().tolist())
                    dpt = dpt  +1 
                    ReElectedLarge.append(pdtempA[pdtempA.NomElu2014 == pdtempB.Nom.unique()[0]])
                    CityLarge.append((d,c))
    print(d,c,dpt)
print(len(ReElectedLarge))
```

```python
dfReelectedLarge = pd.concat(ReElectedLarge, axis=0)
print(len(ReElectedLarge),len(dfReelectedLarge))
dfReelectedLarge.head()
```
